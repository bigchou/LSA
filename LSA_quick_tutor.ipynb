# Score term weights and construct the term-document matrix A and query matrix

```python
import numpy as np

# terms
print("list all terms: ");
terms = [i for line in documents for i in line.lower().split()]
terms = sorted(list(set(terms))) # remove duplicate values and sort list alphabetically
print(terms)

# count frequency of terms
print("\nlist the term-document matrix A...>")
A = np.zeros(shape=(len(terms),len(documents)))
for i,vr in enumerate(terms):
    for j,vc in enumerate(documents):
        for k in vc.lower().split():
            if k == vr:
                A[i][j] += 1
print(A)

# construct query matrix
print("\nlist the query matrix...>")
#query = 'gold silver truck'
#query = "Human computer interaction"
query = "elephant"
q = np.zeros(shape=(len(terms),1))
for i,vr in enumerate(terms):
    for j in query.lower().split():
        if j == vr:
            q[i][0] += 1
print(q)
```

# A = U \* S \* transpose_of_V

```python
U, S, V = np.linalg.svd(A,full_matrices=False)
print("U = ")
print(U)
print("S = ")
print(np.diag(S))
print("V = ")
print(V)
```

# Rank n Approximation

```python
energy = 0.1
sum = np.sum(S)
rank_n = count = 0.0
for i,v in enumerate(sorted(S)):
    count += v
    if count/sum >= energy:
        rank_n = i+1
        break
print("rank_n:",rank_n)

U = np.delete(U,np.s_[rank_n:len(documents)],axis=1)
S = np.delete(S,np.s_[rank_n:len(documents)],axis=0)
V = np.delete(V,np.s_[rank_n:len(documents)],axis=0)
#U = np.delete(U, len(documents)-1, axis=1)
#S = np.delete(S, len(documents)-1, axis=0)
#V = np.delete(V, len(documents)-1, axis=0)
print("U = ")
print(U)
print("S = ")
print(np.diag(S))
print("V = ")
print(V)
```

# Find the new document vector

```python
# document vector
d = []
for i in range(len(documents)):
    d.append([row[i] for row in V])

print("document vector:")
for i,v in enumerate(d):
    print('d'+str(i),v)
```

# Find the new query vector

```python
# find new query vector
print("q = ")
transpose_q = q.transpose()
print(transpose_q)
print("U = ")
print(U)
print("S = ")
print(np.diag(S**-1))
new_q = np.matrix(transpose_q) * U * np.diag(S**-1)
print("new query vector = ")
print(new_q)
```

```python
# compute similarity
sim = {}
for i,v in enumerate(d):
    sim[i] = (np.dot(new_q,v) /(np.linalg.norm(new_q) * np.linalg.norm(v)) )


print("sim =")
for i in sim:
    print(i,sim[i])
```

# Compute cosine similarity

```python
# display result
import operator
sim_sort = sorted(sim.items(),key = operator.itemgetter(1),reverse = True)
print(sim_sort)

for i in sim_sort:
    print(documents[i[0]])
```
